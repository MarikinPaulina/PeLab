```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.linspace(0,0.02)
U1 = 10*np.sin(100*np.pi*t)
U2 = 20*np.sin(200*np.pi*t)
plt.plot(U1,U2,'b',label=r'$10\sin(100\pi t)$')
plt.grid()
plt.xlabel('U1[V]')
plt.ylabel('U2[V]')
plt.savefig('Wykres A-linie.png',dpi=200)
```

```python
U1 = 20*np.sin(100*np.pi*t-np.pi)
U2 = 10*np.sin(200*np.pi*t)
plt.plot(U1,U2,'b',label=r'$10\sin(100\pi t)$')
plt.grid()
plt.xlabel('U1[V]')
plt.ylabel('U2[V]')
plt.savefig('Wykres B-linie.png',dpi=200)
```
